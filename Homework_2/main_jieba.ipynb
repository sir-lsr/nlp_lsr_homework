# 结巴分词实践

## 结巴分词的三种模式

### 精确模式

```python
import jieba

seg = jieba.cut("我来到北京清华大学", cut_all=False)
print("精确模式:", list(seg))
```

### 全模式

```python
seg = jieba.cut("我来到北京清华大学", cut_all=True)
print("全模式:", list(seg))
```

### 搜索引擎模式

```python
seg = jieba.cut_for_search("小明硕士毕业于中国科学院计算所，后在日本京都大学深造")
print("搜索引擎模式:", list(seg))
```

## 载入词典

```python
jieba.load_userdict("userdict.txt")
seg = jieba.cut("李姝蓉即将是自然语言处理方面的高手。", HMM=False)
print("载入词典后分词:", list(seg))
```
