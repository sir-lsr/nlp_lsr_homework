```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import jieba
from gensim.models import Word2Vec

df = pd.read_csv('materials/train.csv')
print(f"数据加载完成，共 {len(df)} 条评论")
```

```python
df['words'] = df['comment'].apply(lambda x: list(jieba.cut(str(x))))
sentences = df['words'].tolist()
print("分词完成，示例:", sentences[0])
```

```python
model_cbow = Word2Vec(sentences, vector_size=100, window=5, min_count=1, sg=0, epochs=10, seed=42)
print("CBOW 模型训练完成")
```

```python
model_sg = Word2Vec(sentences, vector_size=100, window=5, min_count=1, sg=1, epochs=10, seed=42)
print("Skip-Gram 模型训练完成")
```

```python
vec = model_sg.wv['环境']
print("'环境' 的词向量（前10维）:\n", vec[:10])
print("词向量形状:", vec.shape)
```

```python
similar = model_sg.wv.most_similar('好吃', topn=3)
print("与 '好吃' 语义最接近的 3 个词:")
print(similar)
```

```python
sim1 = model_sg.wv.similarity('好吃', '美味')
sim2 = model_sg.wv.similarity('好吃', '蟑螂')
print(f"'好吃' 与 '美味' 的相似度: {sim1:.4f}")
print(f"'好吃' 与 '蟑螂' 的相似度: {sim2:.4f}")
```

```python
result = model_sg.wv.most_similar(positive=['餐厅', '聚会'], negative=['安静'], topn=1)
print("餐厅 + 聚会 - 安静 = ？")
print(result)
```
